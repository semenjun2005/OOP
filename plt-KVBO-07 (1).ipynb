# Опишите что в данном датасете

Данный датасет содержит данные о продажах различных косметических и гигиенических товаров за 12 месяцев. В таблице представлены следующие колонки:

- icecream: Количество проданного мороженого.
- facewash: Количество проданного средства для умывания.
- toothpaste: Количество проданной зубной пасты.
- bathingsoap: Количество проданного мыла для ванны.
- shampoo: Количество проданного шампуня.
- moisturizer: Количество проданного увлажняющего крема.
- totalunits**: Общее количество проданных единиц товаров за месяц.
- **totalprofit: Общая прибыль от продаж за месяц.

Каждая строка соответствует определенному месяцу, начиная с первого месяца (monthnumber = 1) и заканчивая двенадцатым (monthnumber = 12). Данные могут быть использованы для анализа тенденций продаж, выявления сезонных колебаний и оптимизации товарного ассортимента.

# Сравните 2 любых вида товара по количеству продаж в каждый месяц и выведите график с помощью bar chart

```python
months = range(1, 13)
bar_width = 0.35
r1 = np.arange(len(months))
r2 = [x + bar_width for x in r1]

plt.figure(figsize=(10, 5))
plt.bar(r2, df1.bathingsoap.tolist(), color='r', width=bar_width, label='Банный крем')
plt.bar(r1, df1.facecream.tolist(), color='b', width=bar_width, label='Крем для лица')
plt.xlabel('Месяц')
plt.ylabel('Количество продаж')
plt.title('Сравнение количества продаж товара 1 и товара 4 по месяцам')
plt.xticks([r + bar_width/2 for r in range(len(months))], months)
plt.legend()
plt.show()
```

# С помощью круговой диаграммы (pie chart) выведите все продажи за пятый месяц

# С помощью круговой диаграммы (pie chart) выведите все продажи за весь год

```python
plt.pie(df1.sum(axis=1).tolist(), autopct='%1.1f%%')
plt.title('Все продажи за один год')
plt.show()
```

# Выведите временную диаграмму продаж "Shampoo" за год

```python
plt.figure(figsize=(10, 5))
plt.plot(df1.index, df1['shampoo'], linestyle='-')
plt.title('Временную диаграмму продаж "Shampoo" за год')
plt.xlabel('Месяц')
plt.ylabel('Количество продаж')
plt.xticks(df1.index)
plt.show()
```

# Выведите на временной диаграмме сколько заработала компания в каждый месяц
