```python
from collections import *
import re


with open('text.txt', 'r', encoding='utf-8') as file:
    text = file.read()

words1 = text.split()
words_count = Counter(words1)


most_use_words = words_count.most_common(5)
print("5 самых употребляемых слов:", most_use_words)


least_use_words = words_count.most_common()[-5:]
print("5 наименее употребляемых слов:", least_use_words)


french_characters = set("éèêëàâôûùîç")
french_word_count = {word: count for word, count in words_count.items() if any(c in french_characters for c in word)}
most_common_french = Counter(french_word_count).most_common(5)
print("5 самых употребляемых французских слов:", most_common_french)

least_common_french = Counter(french_word_count).most_common()[-5:]
print("5 наименее употребляемых французских слов:", least_common_french)



#3
RussianCount = namedtuple('RussianCount', ['word', 'count'])
named_tuples = [RussianCount(word, count) for word, count in most_use_words]
print("\nИменованные кортежи для самых употребляемых русских слов:")
for item in named_tuples:
    print(item)

FrenchCount = namedtuple('FrenchCount', ['word', 'count'])
named_tuples = [FrenchCount(word, count) for word, count in most_common_french]
print("\nИменованные кортежи для самых употребляемых французских слов:")
for item in named_tuples:
    print(item)


#4
#OrderedDict сохраняет порядок добавления элементов. 
#Это значит, что если вы добавите элементы в определенном порядке, они будут извлекаться в том же порядке.
ordered_dict = Counter({'apple': 3, 'banana': 2, 'cherry': 1})
print("\nOrderedDict:")
for fruit in ordered_dict:
    print(f"{fruit}: {ordered_dict[fruit]}")



#defaultdict позволяет установить значение по умолчанию для отсутствующих ключей. 
#В данном случае мы используем int, что означает, что если ключ не существует, он будет автоматически инициализирован значением 0.
default_dict = defaultdict(int)
default_dict['apple'] += 1
default_dict['banana'] += 2
print("\ndefaultdict:")
for fruit in default_dict:
    print(f"{fruit}: {default_dict[fruit]}")




# UserDict - обертка вокруг обычного словаря с дополнительными методами
class CustomDict(UserDict):
    def __setitem__(self, key, value):
        super().__setitem__(key, value)  # Вызов метода родительского класса
        print(f"Added {key}: {value}")  # Вывод сообщения при добавлении элемента

user_dict = CustomDict()
user_dict['apple'] = 3 
user_dict['banana'] = 2
```
