### Удалите столбцы, где количество утерянных данных превышает 50%

```python
missing_values = (df.isna().sum() * 100 / len(df)).sort_values()
columns_to_drop = missing_values[missing_values > 50].index
df = df.drop(columns=columns_to_drop)
```
